```python
from transformers import pipeline, Conversation
```

```python
conversational_pipeline = pipeline("conversational")
```

```python
tr_list = []
for i in range(len(data)):
  tr1 = "'"+str(data.iloc[i,1]) + "'"
  tr2 = "'" + str(data.iloc[i,0]) + "'"
  tr_list.append(tr1)
  tr_list.append(tr2)
```

```python
answer = []
for i in data1['Question']:
  answer.append(conversational_pipeline([Conversation(i)]))
```

```python
pd.Series(answer).to_csv('mock_answer.csv')
```
